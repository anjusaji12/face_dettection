```python
import cv2
```

```python
img=cv2.imread(r'C:\Users\ANJU\OneDrive\Desktop\DL project\face detection\IMG_9000.JPG')
print(img)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_cascade=cv2.CascadeClassifier(r'C:\Users\ANJU\OneDrive\Desktop\DL project\face detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_gray=cv2.cvtColor(conv_img,cv2.COLOR_BGR2RGB)
img_gray
```

```python
face=face_cascade.detectMultiScale(img_gray,minNeighbors=15)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),color=(0,0,255),thickness=10)
    
```

```python
plt.imshow(conv_img)
```

```python
Video=cv2.VideoCapture(r"C:\Users\ANJU\OneDrive\Desktop\DL project\face detection\video.mp4")
while True:
    success,frame=Video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=22)
    #print(faces)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    cv2.imshow('VIDEO',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
Video.release()
cv2.distroyAllWindows()
```
